## 基础知识
*****************

1. 两个空格或者空行为换行，否则会一直向后延伸下去.
******

2. \为转义符号，可使符号正常显示

*******
3. \# 用来创建标题，\#的数量用来控制标题的大小，\#越多，标题越小

*******
4. 用一个\*号包围的文字为斜体文字，两个是粗体文字，三个是粗斜体文字

    + *斜体文字*

    + **粗体文字**

    + ***粗斜体文字***

*********

5. 三个以上的\*号为创建分割线

*****

6. ~~如果段落上的文字要添加删除线，只需要在文字的两端加上两个波浪线~~
*******

7. 下划线可以通过HTML的标签来实现

    + <u> 此段文字加下划线 </u>
**********

8. 脚注是对文本的补充说明。Markdown 脚注的格式如下:

    + 创建脚注格式类似这样： [^脚注]。

    + [^脚注]: 这是一个脚注 

## 列表的编写
**********
### 无序列表的编写
使用*或-或+
* 列表
+ 列表
- 列表
*************
### 有序列表的编写
使用数字加上\.号来表示，如：
1. 列表
2. 列表
3. 列表
************
### 列表的嵌套
使用TAB键控制列表的嵌套
1. 列表
    1. 列表
        1. 列表
        2. 列表
        3. 列表
    2. 列表
    3. 列表
2. 列表
3. 列表
    


## Markdown区块
*****************
### 区块引用
在段落开头使用 \> 符号，然后段落后紧跟两个空格符号
> 区块引用  
> 区块引用  
> 区块引用
*********************
### 区块嵌套
区块同样可以嵌套，例如：
> 最外层  
>> 第一层嵌套  
>>> 第二层嵌套  
*******************
### 区块中使用列表
> 区块中使用列表  
> 1. 列表  
> 2. 列表  
> 3. 列表  
> * 列表  
> + 列表  
> - 列表  
*************
### 列表中使用区块
同样使用TAB键来控制缩进，记得两个空格
1. 列表
    > 区块  
    > 区块  
2. 列表

## Markdown链接
************************
链接使用方法如下：

\[链接名称\]\(链接地址\)

或者

\<链接地址\>

例如：
这是一个链接[百度](http://www.baidu.com)

或者

<http://www.baidu.com>

## Markdown表格
**************
Markdown使用\|来分隔单元格，并用\-来创建表头

例如：

|姓名|身高|体重|
|---|---|---|
|小明|1.75|140|
|小红|1.65|94|

**************
用\:来控制左右对齐与居中对齐，<u>注意\:必须是英文\:</u>

|姓名|身高|体重|
|:---|:---:|---:|
|小明|1.75|140|
|小红|1.65|94|
